```python
from rotorgraph import *
from particleconfig import *
```

```python
# Tore grid of length nxn

n=4  
n_2 = n*n
```

```python


####graph G_0 : Position + orientation
graph_0= RotorGraph()
#nodes
for i in range(1, 4*n_2 + 1): graph_0.add_node(i)

#Nodes  i*n + j -> case (i,j)

####edges
# N orientation nodes
for i in range (1, n_2 +1):
    l,c = i//n , i%n
    if c==0: l,c = l-1, n 

    
    #Right
    c_r = (c+1)%n 
    if c_r == 0: c_r = n
    i_r = l*n + c_r + 1*n_2

    graph_0.add_edge(i, i_r)

    #Left
    c_l = (c-1)%n
    if c_l == 0: c_l = n
    i_l = l*n + c_l + 3*n_2

    graph_0.add_edge(i, i_l)


# S orientation nodes
for i in range (1+ 2*n_2, 3*n_2 +1):
    l,c = (i-2*n_2)//n , (i-2*n_2)%n
    if c==0: l,c = l-1, n 

    
    #Right
    c_r = (c-1)%n 
    if c_r == 0: c_r = n
    i_r = l*n + c_r + 3*n_2

    graph_0.add_edge(i, i_r)

    #Left
    c_l = (c+1)%n
    if c_l == 0: c_l = n
    i_l = l*n + c_l + 1*n_2

    graph_0.add_edge(i, i_l)


# E orientation nodes
for i in range (1+ n_2, 2*n_2 +1):
    l,c = (i-n_2)//n , (i-n_2)%n
    if c==0: l,c = l-1, n 

    
    #Right
    l_r = (l-1)%n 
    i_r = l_r*n + c  + 2*n_2

    graph_0.add_edge(i, i_r)

    #Left
    l_l = (l+1)%n 
    i_l = l_l*n + c 

    graph_0.add_edge(i, i_l)


# O orientation nodes
for i in range (1+ 3*n_2, 4*n_2 +1):
    l,c = (i-3*n_2)//n , (i-3*n_2)%n
    if c==0: l,c = l-1, n 

    
    #Right
    l_r = (l+1)%n 
    i_r = l_r*n + c 

    graph_0.add_edge(i, i_r)

    #Left
    l_l = (l-1)%n 
    i_l = l_l*n + c + 2*n_2

    graph_0.add_edge(i, i_l)


```

```python


####graph G_1: Color of the position
graph_1= RotorGraph()
#nodes
for i in range(1, 2*n_2 + 1): graph_1.add_node(i)
#edges
for i in range(1, n_2+1):
    graph_1.add_edge(i, n_2+i)
    graph_1.add_edge(n_2 +i, i)

```

```python



###Update rule - Legale routing

#For one ant:
#initial config
node= 1
sigma = ParticleConfig(graph_1)

#All white (R) color 
for i in range(1,n_2+1):
    sigma[i] = 1


step =5
visited_nodes = [node]

while step>0:
    #cel of particle.
    cel = node % n_2
    if cel==0: cel=n_2

    #Mouvement of the ant according to the color of the cel + cel color update
    if sigma[cel]==1:
        node = (graph_0.rotor_order[node][0])[1]

        sigma[cel] = 0
        sigma[cel + n_2] = 1
        
    else :
        node = (graph_0.rotor_order[node][1])[1]

        sigma[cel]=1
        sigma[cel+ n_2]= 0



    visited_nodes.append(node) # update of the viseted nodes
    step -=1 #step update


# test 
print(visited_nodes)
print(sigma)
# print(graph_0.edges())
# print(graph_0.rotor_order[12][0])
```
